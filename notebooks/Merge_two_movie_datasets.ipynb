```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
movie_data=pd.read_csv("../data/final_data.csv", 
               usecols = ['id', 'Name', 'PosterLink', 'Genres', 'Actors', 
                          'Director','Description', 'DatePublished', 'Keywords'], 
               parse_dates = ["DatePublished"])
movie_data["year"] = movie_data["DatePublished"].dt.year.fillna(0).astype(int)
movie_data= movie_data.drop(["DatePublished"], axis=1, inplace=True)
movie_data
```

```python
plots_data = pd.read_csv('../data/wiki_movie_plots_deduped.csv')
plots_data = plots_data[plots_data['Release Year'] > 1970]
plots_data = plots_data[plots_data.duplicated(subset=['Title', 'Release Year', 'Plot']) == False]
plots_data = plots_data[plots_data.duplicated(subset=['Title', 'Release Year']) == False]
plots_data = plots_data[['Title', 'Plot', 'Release Year']]

plots_data.columns = ['Name', 'Plot', 'year']
plots_data
```

```python
movie_data = movie_data.merge(plots_data, on=['Name', 'year'], how='left').fillna('')
movie_data.reset_index(drop=True, inplace=True)
movie_data
```

```python
movie_data.to_csv("../data/movie_data_merged.csv",index=False)
pd.read_csv("../data/movie_data_merged.csv",index_col='id')
```
